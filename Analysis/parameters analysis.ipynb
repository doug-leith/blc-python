```python
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.cm as cm
```

```python
df = pd.read_csv('output_big.csv')
```

# Analysis of Movielens 20m using 10% test set
We are using Movielens 20m to be comparable with other papers (24m is too recent)

```python
with pd.option_context('display.max_rows', 999, 'display.max_columns', 6):
    print(df[['p','sigma','d','Prediction_RMSE','Time']])
```

## RMSE Analysis (we use first 75 elements to show the trend to get uniform distribution across parameters)

```python
df.iloc[0:75].boxplot(by='p',column=['Prediction_RMSE']);
```

```python
df.iloc[0:75].boxplot(by='sigma',column='Prediction_RMSE');
```

```python
df.iloc[0:75].boxplot(by='d',column='Prediction_RMSE');
```

### Best results:

```python
# best parameters
df.loc[df['Prediction_RMSE'].argmin()]
```

```python
fig = plt.figure()
ax = Axes3D(fig)
p = ax.plot_trisurf(df.p, df.d, df.Prediction_RMSE, cmap=cm.jet, linewidth=0.2)
fig.colorbar(p)
plt.xlabel("p")
plt.ylabel("d")
plt.title("Prediction RMSE over p and d")
fig = plt.figure()
ax = Axes3D(fig)
p = ax.plot_trisurf(df.p, df.sigma, df.Prediction_RMSE, cmap=cm.jet, linewidth=0.2)
fig.colorbar(p)
plt.xlabel("p")
plt.ylabel("sigma")
plt.title("Prediction RMSE over p and sigma")
fig = plt.figure()
ax = Axes3D(fig)
p = ax.plot_trisurf(df.d, df.sigma, df.Prediction_RMSE, cmap=cm.jet, linewidth=0.2)
fig.colorbar(p)
plt.xlabel("d")
plt.ylabel("sigma")
plt.title("Prediction RMSE over d and sigma")
pass
#ax.set_zlim3d(0,2)
```

## Time analysis

```python
#Time
df.iloc[0:75].boxplot(by='p',column=['Time']);
df.iloc[0:75].boxplot(by='sigma',column='Time');
df.iloc[0:75].boxplot(by='d',column='Time');
```

```python
fig = plt.figure()
ax = Axes3D(fig)
p = ax.plot_trisurf(df.p, df.d, df.Time, cmap=cm.jet, linewidth=0.2)
fig.colorbar(p)
plt.xlabel("p")
plt.ylabel("d")
plt.title("Time over p and d")
fig = plt.figure()
ax = Axes3D(fig)
p = ax.plot_trisurf(df.p, df.sigma, df.Time, cmap=cm.jet, linewidth=0.2)
fig.colorbar(p)
plt.xlabel("p")
plt.ylabel("sigma")
plt.title("Time over p and sigma")
fig = plt.figure()
ax = Axes3D(fig)
p = ax.plot_trisurf(df.d, df.sigma, df.Time, cmap=cm.jet, linewidth=0.2)
fig.colorbar(p)
plt.xlabel("d")
plt.ylabel("sigma")
plt.title("Time over d and sigma")
pass
```
